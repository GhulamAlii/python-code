```python

img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')

def nothing(x): 
    pass

cv2.createTrackbar('color1','image',5,255,nothing)
cv2.createTrackbar('color2','image',5,255,nothing)
cv2.createTrackbar('color3','image',5,255,nothing)
# cv2.createTrackbar('thickness','image',-1,5,nothing)
def draw_circle(event,x,y,flags,param): 
    if event == cv2.EVENT_LBUTTONDBLCLK: 
        cv2.circle(img,(x,y),100,(color1,color2,color3),-1)

cv2.setMouseCallback('image',draw_circle)

while True:
    
    cv2.imshow('image',img)
    
#   radius = cv2.getTrackbarPos('Radius','image')
    color1 = cv2.getTrackbarPos("color1",'image')
    color2 = cv2.getTrackbarPos("color2",'image')
    color3 = cv2.getTrackbarPos("color3",'image')
    
    
    if cv2.waitKey(20) & 0xFF == ord('c') :
        img = np.ones((512,512,3), np.uint8)
    
    
    if cv2.waitKey(20) & 0xFF == 27 or cv2.getWindowProperty('image',cv2.WND_PROP_VISIBLE) < 1: 
        break
           
            
cv2.destroyAllWindows()
```
