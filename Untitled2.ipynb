```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import mediapipe as mp
from time import time
```

```python
hand_sol = mp.solutions.hands

hand_func = hand_sol.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.3)
draw = mp.solutions.drawing_utils
```

```python
img = cv2.imread("img/ali.jpg")
plt.imshow(img[...,::-1])
```

```python
result = hand_func.process(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

if result.multi_hand_landmarks:
    for hand_no, hand_landmarks in enumerate(result.multi_hand_landmarks):
        print(f"hand number: {hand_no+1}")
        print('-------------')
        
        for i in range(2):
            print(f"{hand_sol.HandLandmark(i).name}")
            print(f"{hand_landmarks.landmark[hand_sol.HandLandmark(i).value]}")
```

```python
a = img.copy()
if result.multi_hand_landmarks:
    for hand_no, hand_landmarks in enumerate(result.multi_hand_landmarks):
        draw.draw_landmarks(image=a,landmark_list = hand_landmarks,
                              connections = hand_sol.HAND_CONNECTIONS)
        
Fig = plt.figure(figsize=[10,10])
plt.imshow(a[...,::-1])
```

```python
def handLandmark(img,hands, display=True):
    
    output = img.copy()
    result = hand_func.process(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
    
    if result.multi_hand_landmarks:
        for hand_no, hand_landmarks in enumerate(result.multi_hand_landmarks):
            draw.draw_landmarks(image=output,landmark_list = hand_landmarks,
                                  connections = hand_sol.HAND_CONNECTIONS)
    
    if display:
        fig = plt.figure(figsize=[10,10])
        plt.imshow(output[...,::-1])
    else:
        return output, result
handLandmark(a,2,display=True)
```

```python
h_vid = hand_sol.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.7, min_tracking_confidence=0.4)


cap = cv2.VideoCapture('http://192.168.1.103:8080/video')
#cap = cv2.VideoCapture('E:\WishDragon.2021.HDRip.Dual.mp4')
cap.set(3,1280)
cap.set(4,960)
while (True):
    ret,frame = cap.read()
    
    if not ret:
        continue
        
    frame = cv2.flip(frame,1)
    
    frame, bl = handLandmark(frame,h_vid, display=False)
    cv2.imshow("frame",frame)
    
    if cv2.waitKey(1) & 0xff == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
